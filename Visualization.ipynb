```python
data_vs['sex'] = data_vs['sex'].replace(np.nan, 0)
data_vs['bill_length_mm']=data_vs["bill_length_mm"].replace(np.nan, 0)
data_vs['bill_depth_mm']=data_vs["bill_depth_mm"].replace(np.nan, 0)
data_vs['flipper_length_mm']=data_vs["flipper_length_mm"].replace(np.nan, 0)
data_vs['body_mass_g']=data_vs["body_mass_g"].replace(np.nan, 0)
```

# **Data Visualization**

```python
numerical_float64 = (data_vs.dtypes == "float64")
numerical_float64_list = list(numerical_float64[numerical_float64].index)

print("Numerical variables:")
print(numerical_float64_list)
```

```python
def plot_hist(variable):
    plt.figure(figsize = (9,3))
    plt.hist(data_vs[variable], bins = 50)
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} Distribution with Histogram".format(variable))
    plt.show()
```

```python
for n in numerical_float64_list:
    plot_hist(n)
```

```python
cols = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
sns.pairplot(data_vs[cols])
```

```python
sns.pairplot(data_vs, hue = 'species')
```

```python
plt.figure(figsize=(12,8)) 
sns.heatmap(data_vs.corr(), annot=True, cmap='Dark2_r', linewidths = 2)
plt.show()
```

```python
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
sns.barplot(x = 'species', y = 'bill_length_mm', data = data_vs, palette="cubehelix")
plt.subplot(2,2,2)
sns.barplot(x = 'species', y = 'bill_depth_mm', data = data_vs, palette="Oranges")
plt.subplot(2,2,3)
sns.barplot(x = 'species', y = 'flipper_length_mm', data = data_vs, palette="Oranges")
plt.subplot(2,2,4)
sns.barplot(x = 'species', y = 'body_mass_g', data = data_vs, palette="cubehelix")
```

```python
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
sns.violinplot(x = 'species', y = 'bill_length_mm', data = data_vs, palette="rocket_r")
plt.subplot(2,2,2)
sns.violinplot(x = 'species', y = 'bill_depth_mm', data = data_vs, palette="rocket_r")
plt.subplot(2,2,3)
sns.violinplot(x = 'species', y = 'flipper_length_mm', data = data_vs, palette="rocket_r")
plt.subplot(2,2,4)
sns.violinplot(x = 'species', y = 'body_mass_g', data = data_vs, palette="rocket_r")
```

```python
plt.figure(figsize=(20,15))
plt.subplot(2,2,1)
sns.boxplot(x = 'species', y = 'bill_length_mm', data = data_vs, palette="gist_ncar_r")
plt.subplot(2,2,2)
sns.boxplot(x = 'species', y = 'bill_depth_mm', data = data_vs, palette="gist_ncar_r")
plt.subplot(2,2,3)
sns.boxplot(x = 'species', y = 'flipper_length_mm', data = data_vs, palette="gist_ncar_r")
plt.subplot(2,2,4)
sns.boxplot(x = 'species', y = 'body_mass_g', data = data_vs, palette="gist_ncar_r")
```
